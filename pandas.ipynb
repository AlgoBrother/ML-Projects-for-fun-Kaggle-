## Remember that if an index is not present in the Data Frame then pandas will automatically create its own index in the Data Frame. It wont throw an error.

```python
new_df.loc[0,0] = 0.506043
new_df
```

## Now that you seen it, let's delete the new index since its irrelevant for us.

```python
new_df.drop(0) # Can use new_df.drop[0] as well
```

## Without using __axis = 1__ it will only delete the row, if you want to delete the column as well, then...

```python
new_df = new_df.drop(0, axis = 1) # new_df.drop(['A'], axis = 1) if you want to delete a named column
new_df
```

## Keep in mind that __axis = 0__ refers to the row and __axis = 1__ refers to the column, [also axis only takes 0 or 1 as input, any other input will raise ValueError] so when you write __drop(0, axis = 0)__ then it means to delete the whole row with index 0. Similarly for __drop(0, axis = 1)__ it means delete the column with index 0.
